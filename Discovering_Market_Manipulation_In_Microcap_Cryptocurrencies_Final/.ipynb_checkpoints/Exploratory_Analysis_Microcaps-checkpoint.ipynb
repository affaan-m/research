**Step 3 - Data Preprocessing:**

    We begin by loading the collected Reddit posts from the CSV file using the pandas library. To focus on recent data, we convert the ‘Date’ column of our dataset to datetime format and filter out posts from the year 2023. This is done using a boolean mask that isolates entries from this year. To ensure that our data is aligned with the time granularity we are interested in, we round the time of each post to the nearest 15 minutes. We then group the posts by the rounded time, as well as by their content and title, and sum up the upvotes for posts falling within each group. This aggregation helps us measure the overall impact and reach of the posts. As a final preprocessing step, we clean the text in each post. This involves removing punctuation, converting text to lowercase, tokenizing the text into individual words, and lemmatizing the words (i.e., reducing them to their base or dictionary form). Furthermore, we remove common words, known as “stop words”, that are often deemed irrelevant for analysis due to their high frequency (e.g., “the”, “and”, “is”). By the end of this step, we have a preprocessed and aggregated dataset ready for sentiment analysis. The processed text, along with the other relevant details of each post, is now ready for the next phase of our study. 

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
from nltk.stem import WordNetLemmatizer
import re
import datetime
import time

# Load the reddit data 
reddit_posts = pd.read_csv('all_posts.csv')

# Load the microcaps data
microcaps = pd.read_excel('microcaps.xlsx')

# Get the CMC IDs as a list
cmc_id_list = microcaps['id'].tolist()

# Convert the 'Date' column to datetime
reddit_posts['Date'] = pd.to_datetime(reddit_posts['Date'])

# Create a boolean mask for the year 2023
mask = (reddit_posts['Date'].dt.year == 2023)

# Apply the mask to the DataFrame
reddit_posts = reddit_posts.loc[mask]

# Round the time to the nearest 15 minutes
reddit_posts['time_rounded'] = reddit_posts['Date'].dt.round('15min')

# Group by the rounded time and the post text, and aggregate the upvotes
reddit_posts_agg = reddit_posts.groupby(['time_rounded', 'Content', 'Title']).agg({'Ups': 'sum'}).reset_index()

stop_words = set(stopwords.words('english')) 
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    text = re.sub(r'[^\w\s]', '', text)  # remove punctuations
    text = text.lower()  # convert to lowercase
    tokenized_text = word_tokenize(text)  # tokenization
    text = [lemmatizer.lemmatize(word) for word in tokenized_text if word not in stop_words]  # lemmatization and stopwords removal
    return text

reddit_posts_agg['processed_text'] = reddit_posts_agg['Content'].apply(preprocess_text)

# Get the microcap tickers as a list
microcap_tickers = microcaps['symbol'].tolist()

# Create a dictionary to store the post counts for each ticker
tickers_with_posts = {}

# Iterate through the tickers and count the number of posts for each ticker
for ticker in microcap_tickers:
    mask = reddit_posts_agg['Content'].str.contains(ticker, case=False) | reddit_posts_agg['Title'].str.contains(ticker, case=False)
    post_count = reddit_posts_agg[mask].shape[0]
    if post_count > 1:
        tickers_with_posts[ticker] = post_count

# Print the dictionary mapping tickers to post counts in descending order
print("Ticker - Number of Posts")
sorted_posts = sorted(tickers_with_posts.items(), key=lambda x: x[1], reverse=True)
for ticker, count in sorted_posts:
    print(f"{ticker}: {count}")

# Print the count of tickers
ticker_count = len(tickers_with_posts)
print(f"Total number of microcaps: {ticker_count}")

# Create a dictionary mapping from ticker to id for all tickers
ticker_to_id_all = dict(zip(microcap_tickers, cmc_id_list))

# Create a dictionary mapping from ticker to id for the tickers in tickers_with_posts
ticker_to_id = {ticker: ticker_to_id_all[ticker] for ticker in tickers_with_posts if ticker in ticker_to_id_all}
```

    After filtering out cryptocurrencies with under 2 posts, our sample size was reduced to 25 microcaps. The post count also decreased due to the aggregation techniques we employed. We print a dictionary that maps each microcap ticker to its post count in descending order. This gives us a sense of which microcap cryptocurrencies are being discussed the most on Reddit. We then create a dictionary from this filtered set that maps the tickers to their respective CMC ID's. This step is done to collect price data from CoinMarketCap's API since we need to have the tickers connected to their respective CMC ID's. 


**Step 4 - Sentiment Analysis:**

    With our data preprocessed, we move on to the sentiment analysis phase. The goal here is to determine the sentiment of each Reddit post in our dataset. To achieve this, we use the VADER sentiment analysis tool from the nltk library. VADER (Valence Aware Dictionary and sEntiment Reasoner) is a lexicon and rule-based sentiment analysis tool that is specifically designed to work well with social media text. We create a SentimentIntensityAnalyzer object, and define a function ‘get_sentiment’ that takes in a text and returns the compound sentiment score. This compound score is a metric that calculates the sum of all the lexicon ratings which have been normalized between -1(most extreme negative) and +1 (most extreme positive). We then apply this function to the content of each Reddit post in our dataset and create a new column ‘sentiment’ to store the sentiment scores. We use descriptive statistics to understand the distribution of sentiment scores in our dataset. 


```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()

def get_sentiment(text):
    sentiment = sia.polarity_scores(text)
    return sentiment['compound']

reddit_posts_agg['sentiment'] = reddit_posts_agg['Content'].apply(get_sentiment)
reddit_posts_agg['sentiment'].describe()
```

    In our sentiment analysis, we found that the compound sentiment was overwhelmingly positive on a scale from -1 to 1. As seen in the summary statistics above. This provided some insight into our theory that there is targeted promotion and shilling of these cryptocurrencies, often motivated by hidden incentives. The reasons range from developer-led pump and dump schemes to insider groups profiting from tiered rankings that delay the release of insider information based on subscription level within the group. Such activities are commonly orchestrated through communities on platforms like Telegram and Discord.

    Before moving on to our next step, we scrape the CMC API for price data on all of our 25 microcap cryptocurrencies over the past 6 months. We take a quote every 15 minute interval to match the aggregation of reddit posts we did which was also on a 15 minute interval. 

```python
import requests
import json
import datetime
import pandas as pd

# Define your CoinMarketCap API key
cmc_API_KEY = 'you api key'

# Define the base URL for the CoinMarketCap API
cmc_url = 'https://pro-api.coinmarketcap.com/v3/cryptocurrency/quotes/historical'

# Define the headers for the API request
headers = {
    'Accepts': 'application/json',
    'X-CMC_PRO_API_KEY': cmc_API_KEY,
}

def get_historical_price(cmc_id, year):
    # Define the time ranges for each month from January to June
    time_ranges = [
        (datetime.date(year, month, 1).isoformat(), 
         datetime.date(year, month+1, 1).isoformat() if month < 6 else datetime.date(year, month, 30).isoformat())
        for month in range(1, 7)
    ]

    all_quotes = []
    for time_start, time_end in time_ranges:
        parameters = {
            'id': cmc_id,
            'time_start': time_start,
            'time_end': time_end,
            'interval': '15m',
            'convert': 'USD',
            'aux': 'price,volume,market_cap,circulating_supply,total_supply,quote_timestamp,is_active,is_fiat',
            'skip_invalid': True
        }

        response = requests.get(cmc_url, headers=headers, params=parameters)

        # Check if the request was successful
        if response.status_code != 200:
            print(f'Request for ID {cmc_id} not successful. Status code: {response.status_code}')
            print(f'Response: {response.text}')
            continue

        response_json = response.json()
        for item in response_json['data'][str(cmc_id)]['quotes']:
            quote = {
                'id': cmc_id,
                'timestamp': item.get('timestamp'),
                'price': item['quote']['USD'].get('price', None),
                'volume_24hr': item['quote']['USD'].get('volume_24hr', None),
                'market_cap': item['quote']['USD'].get('market_cap', None),
                'conversion_timestamp': item['quote']['USD'].get('timestamp', None),
            }
            all_quotes.append(quote)

    return all_quotes

all_dataframes = []
for ticker, cmc_id in ticker_to_id.items():
    quotes = get_historical_price(cmc_id, 2023)  # Replace 2023 with the year you want
    df_price = pd.DataFrame(quotes)

    # Convert the 'timestamp' to datetime and set it as the index
    df_price['timestamp'] = pd.to_datetime(df_price['timestamp'])
    df_price.set_index('timestamp', inplace=True)
    
    all_dataframes.append(df_price)

# Concatenate all the dataframes into a single dataframe
df_all = pd.concat(all_dataframes, ignore_index=True)

# save our price data as a csv file
df_all.to_csv('df_all.csv')

```

**Step 5 - Correlation Analysis:**

    We calculate the correlation between the number of Reddit posts (a measure of discussion volume), and the price of each cryptocurrency. This provides a measure of how closely these variables move together. If the correlation is positive, it means that when the number of Reddit posts increases, the price of the cryptocurrency tends to increase as well, and vice versa. If the correlation is negative, it means that when the number of Reddit posts increases, the price of the cryptocurrency tends to decrease, and vice versa. A correlation close to zero indicates that there is no linear relationship between the variables.


```python
import warnings
from scipy import stats
warnings.filterwarnings('ignore')

# Create a DataFrame to store the correlation results
correlation_df = pd.DataFrame(index=list(tickers_with_posts.keys()), columns=['Correlation', 'P-value'])

# Iterate through the tickers with posts and perform correlation analysis
for ticker in tickers_with_posts.keys():
    try:
        # Filter posts based on the ticker
        mask = reddit_posts_agg['Content'].str.contains(ticker, case=False) | reddit_posts_agg['Title'].str.contains(ticker, case=False)
        ticker_df = reddit_posts_agg[mask]

        # Retrieve historical price data for the ticker
        ticker_id = ticker_to_id.get(ticker)
        if ticker_id is None:
            raise KeyError(f"Ticker ID not found for ticker: {ticker}")
        ticker_price_df = df_all[df_all['id'] == ticker_id]

        # Convert conversion_timestamp column to datetime64[ns]
        ticker_price_df.loc[:, 'conversion_timestamp'] = pd.to_datetime(ticker_price_df['conversion_timestamp'])

        # Convert the conversion_timestamp column to UTC datetime format
        ticker_price_df.loc[:, 'conversion_timestamp'] = ticker_price_df['conversion_timestamp'].dt.tz_convert('UTC')

        # Convert the time_rounded column to UTC datetime format
        ticker_df.loc[:, 'time_rounded'] = pd.to_datetime(ticker_df['time_rounded']).dt.tz_localize('UTC')

        # Merge the price and posts data on the common timestamp
        merged_df = pd.merge(ticker_price_df, ticker_df, left_on='conversion_timestamp', right_on='time_rounded', how='inner')

        # Calculate the correlation coefficient and p-value
        correlation_coefficient, p_value = stats.pearsonr(merged_df['price'], merged_df['Ups'])

        # Store the correlation coefficient and p-value in the DataFrame
        correlation_df.loc[ticker, 'Correlation'] = correlation_coefficient
        correlation_df.loc[ticker, 'P-value'] = p_value

    except KeyError as e:
        print(f"KeyError: {e} for ticker: {ticker}")
    except ValueError as e:
        print(f"ValueError: {e} for ticker: {ticker}")

# Sort the correlation results by correlation coefficient in descending order
correlation_df.sort_values(by='Correlation', ascending=False, inplace=True)

# Print the correlation results
print(correlation_df)
```

    Above we just have the correlation and the p - values associated to test whether there is a statistically significant correlation or not. Below we add the post count to better understand how the post count might relate to the correlation as well as drop all null microcaps that didn't have enough posts to produce results.
    
**Step 6 - Statistical Significance Testing:**

    In this step, we test whether the relationships we have found in the data are statistically significant. This means checking if the relationships are unlikely to have arisen by chance. We do this by calculating p-values. If the p-value is small (typically, less than 0.05), we conclude that the relationship is statistically significant. For example, if the p-value of the correlation between Reddit discussions and cryptocurrency prices is small, we can confidently say that there is a significant correlation between these two variables.

```python
# Drop the null values
correlation_df.dropna(inplace=True)

# Add a new column 'Post Count' to the DataFrame
correlation_df['Post Count'] = correlation_df.index.map(tickers_with_posts)

# Check the DataFrame
print(correlation_df)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(correlation_df['Correlation'], correlation_df['P-value'], alpha=0.5, s=correlation_df['Post Count'])
plt.xlabel('Correlation')
plt.ylabel('P-value')
plt.title('Scatter plot of Correlation and P-value')
plt.grid(True)
plt.show()
```

```python
# Create a mask where 'P-value' is less than 0.05
mask = correlation_df['P-value'] < 0.05

# Use the mask to subset the DataFrame
correlation_df_95 = correlation_df.loc[mask]
print(correlation_df_95)
```

    Above, we conducted statistical testing for correlations. The size of the circles in the visualrepresentation above corresponds to the post count for each cryptocurrency. We used a 95% confidence interval, as is common in most academic studies. Only two of our microcaps met this criterion and both showed strong positive correlations: as is seen in the table above. Both can be visually identified in the bottom right corner of the plot above.

## Evidence of Market Manipulation:

    If a coin is initially unknown and not posted anywhere outside of a Reddit forum or a few tweets, then the information likely originates from two sources. This could also involve a "rug pull", a term used to describe cryptocurrency founders or developers who leave the token liquidity unlocked. This action allows them to pull all funds out of the market, rendering the currency worthless.
